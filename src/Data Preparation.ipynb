```python
import pandas as pd
import numpy as np
```

```python
doc_id = ""
sent_no = 7520
data_dict = {"Document ID":[],"Word":[],"Sentence No":[],"Tag":[]}
with open("pubmed.txt","r") as f:
    for line in f.readlines():
        if "doc_id" in line:
            doc_id = line.split(" = ")[-1][:-1]
        else:
            temp = line.split()
            if temp:                
                data_dict["Document ID"].append(doc_id)
                data_dict["Sentence No"].append(sent_no)
                data_dict["Tag"].append(temp[1])
                data_dict["Word"].append(temp[0])
                if(temp[0] == "."):
                    sent_no = sent_no + 1

data = pd.DataFrame(data_dict)
data = data[["Document ID","Sentence No","Word","Tag"]]
data
```

```python
data.to_csv("PubMed.csv", encoding='utf-8')
```
