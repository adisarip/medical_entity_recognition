<h2 style="color:#081d57"> Creating .ann file from micromed dataset</h2>

```python
# listner = stdOutListner()
auth = tweepy.OAuthHandler( twitter_credentials.CONSUMER_KEY, twitter_credentials.CONSUMER_SECRET )
auth.set_access_token( twitter_credentials.ACCESS_TOKEN, twitter_credentials.ACCESS_TOKEN_SECRET )
   
api = tweepy.API(auth)
# tweets = open("tweets_2.txt", "w+")
avail_tweet_ids = []
for tid in tweet_ids :
    try :
        tweet = api.get_status(tid)
        avail_tweet_ids.append(tid)
#         tweets.write(tid + " : " + tweet.text + " \n" )
        ff = open("./Twitter_370_txt/"+tid + ".txt", "w+")
        ff.write(tweet.text)
        ff.close()
        txt = tweet.text
        to_write = ""
        
        fd = open("micromed/medinfo2015.linejson", "r")
        # fdw = open("Tweet_ids.txt", "w+")
        id_found = False
        for line in fd.readlines() :
            data = json.loads(line)
            if data["tid"] == tid :
                id_found = True
                if len(data["entities"]) > 0 :
                    i = 0 
                    while i < len(data["entities"]) :
                        tag = data["entities"][i]["type"]
                        if tag.strip() == "Pharmacological_Substance" :
                            tag = "Drug"
                        l = 0 
                        while l < len(data["entities"][i]["locations"]) :
                            start, end = data["entities"][i]["locations"][l]["start"], data["entities"][i]["locations"][l]["end"]
                            word = txt[start:end]
                            to_write += ("T" + str(i + l+1) + "	" + tag + " " + str(start) + " " + str(end) + "	"+word + "\n")
                            l += 1
                
                        i += 1
                ff = open("./Twitter_370_ann/"+tid+".ann", "w+")
                ff.write(to_write)
                ff.close()
            if id_found :
                break
        
        fd.close()
        
    except tweepy.TweepError :
        print("Error in tweet id : ", tid)
    
# tweets.close()
# tweet_csv = [ tweet.text for tweet in tweets ]
# temp = []
# for j in tweets :
#     temp.append( j )

# print( temp )    
```

<h1 style="color:#0a6313"> Script to check duplicate data generated by [standoff2conll.py](
https://github.com/spyysalo/standoff2conll/blob/master/standoff2conll.py) scipt, and create a csv file from generated text file by [standoff2conll.py](
https://github.com/spyysalo/standoff2conll/blob/master/standoff2conll.py)  </h1>

```python
file = "./standoff2conll-master/data_twitter_521.txt"
ff = open(file, "r")
cs = []
doc = 0
count = 0
docs = []
l_no = 0
lines = ff.readlines()
i = 0
while i < len(lines) :
    line = lines[i]
    l_no += 1
    if len(line.strip()) > 0 :
#         line = line.split()
#         print(line)
        if "# doc_id = " in line :
            count += 1
            doc = line[ len("# doc_id = ") :].strip()
            if doc in docs :
                while doc in docs :
                    print("Duplicate doc found!!! : ", doc)
                    print("Line number : ", i )
                    i += 1
                    while i < len(lines) :
#                         print("i : ", i)
                        line = lines[i]
                        if "# doc_id = " in line :
                            doc = line[ len("# doc_id = ") :].strip()
                            break
#                         print("Before update : ", i)
                        i += 1
#                         print("After update : ", i)
#                 print("[REPEAT] : ", l_no, "::", doc)
            docs.append(doc)
            print("Doc id : ", doc )
            
        else :
            line = line.split()
            line = [doc] + line
            cs.append(line)
    i += 1
```

```python
import pandas as pd

data = pd.DataFrame(cs)
data.to_csv("twitter_521.csv", index=False)
```
